```python
import pandas as pd
import numpy as np

# โหลดข้อมูล (สมมติว่าอ่านจากไฟล์แล้ว)
# df = pd.read_csv('/path/to/risk_Cervicalupdated1.csv')

# ฟังก์ชันช่วยเติมข้อมูลตามเงื่อนไข
def fill_missing_values(df):
    # ดึงค่าที่ไม่เป็น NaN จากคอลัมน์ 'First sexual intercourse' เพื่อใช้สำหรับการสุ่ม
    non_null_values = df['First sexual intercourse'].dropna()

    for index, row in df.iterrows():
        age = row['Age']

        # 1. Number of sexual partners (เริ่มมีได้เมื่ออายุ 15+)
        if pd.isna(row['Number of sexual partners']):
            if age < 15:
                df.at[index, 'Number of sexual partners'] = 0
            else:
                df.at[index, 'Number of sexual partners'] = np.random.randint(1, min(10, age - 14))

        # 2. First sexual intercourse (อิงจากอายุและจำนวน sexual partners)
        if pd.isna(row['First sexual intercourse']):
            # ใช้สุ่มค่าจาก non_null_values แทนการใช้การคำนวณตามอายุ
            df.at[index, 'First sexual intercourse'] = np.random.choice(non_null_values)

        # 3. Num of pregnancies (สัมพันธ์กับอายุและ partners)
        if pd.isna(row['Num of pregnancies']):
            if age < 18:
                df.at[index, 'Num of pregnancies'] = 0
            else:
                # ดัก NaN ก่อน
                if pd.isna(row['Number of sexual partners']):
                    num_partners = 0
                else:
                    num_partners = int(row['Number of sexual partners'])
                    df.at[index, 'Num of pregnancies'] = np.random.randint(0, min(11, num_partners + 2))

        # 4. Smokes (สัมพันธ์กับ Smokes (years))
        if pd.isna(row['Smokes']):
            if pd.isna(row['Smokes (years)']) or row['Smokes (years)'] == 0:
                df.at[index, 'Smokes'] = 0
            else:
                df.at[index, 'Smokes'] = 1

        # 5. Smokes (years) (สัมพันธ์กับ Smokes)
        if pd.isna(row['Smokes (years)']):
            if row['Smokes'] == 0:
                df.at[index, 'Smokes (years)'] = 0
            else:
                # สมมติว่าถ้าเริ่มสูบบุหรี่ จะเริ่มสูบช่วงวัยรุ่น-วัยทำงาน
                max_smoke_years = max(0, age - 14)
                df.at[index, 'Smokes (years)'] = np.random.uniform(0.5, max_smoke_years)

    return df

# สมมติว่า df คือ DataFrame ที่โหลดมาแล้ว
# เรียกใช้งานฟังก์ชันเติมข้อมูล
df_filled = fill_missing_values(df)

# ตรวจสอบว่าข้อมูล NaN หายไปหรือไม่
print(df_filled.isna().sum())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# โหลด Dataframe ของคุณ (ในที่นี้สมมติว่า df คือ DataFrame ของคุณที่ทำไว้แล้ว)

# ฟังก์ชันสำหรับเช็ค Skewness และ Plot กราฟ
def check_skewness_and_plot(df, numeric_columns):
    skewness_result = []

    plt.figure(figsize=(15, 60))

    for i, col in enumerate(numeric_columns):
        # คำนวณ Skewness
        col_skewness = skew(df[col].dropna())

        # จัดกลุ่มว่าเบ้ขวาหรือซ้าย หรือสมมาตร
        if col_skewness < -1:
            status = "Highly Negatively Skewed (เบ้ซ้ายมาก)"
        elif col_skewness < -0.5:
            status = "Moderately Negatively Skewed (เบ้ซ้ายเล็กน้อย)"
        elif col_skewness < 0.5:
            status = "Symmetrical (สมมาตร)"
        elif col_skewness < 1:
            status = "Moderately Positively Skewed (เบ้ขวาเล็กน้อย)"
        else:
            status = "Highly Positively Skewed (เบ้ขวามาก)"

        skewness_result.append([col, col_skewness, status])

        # Plot Histogram + KDE Plot
        plt.subplot(len(numeric_columns), 2, i+1)
        sns.histplot(df[col], kde=True, color='skyblue')
        plt.title(f"{col}\nSkewness: {col_skewness:.4f} - {status}")

    plt.tight_layout()
    plt.show()

    # แสดงตารางสรุป Skewness
    skewness_df = pd.DataFrame(skewness_result, columns=['Column', 'Skewness', 'Status'])
    return skewness_df

# ระบุเฉพาะคอลัมน์ที่เป็นตัวเลขเท่านั้น
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# เรียกฟังก์ชันเช็ค Skewness + Plot
skewness_summary = check_skewness_and_plot(df, numeric_columns)

# ดูสรุปค่า Skewness
print(skewness_summary)

```

```python
import pandas as pd
import numpy as np
from scipy.stats import yeojohnson, boxcox, skew

# Backup ข้อมูลก่อนแปลง เก็บเป็น df_backup
df_backup = df.copy()

# ฟังก์ชันช่วยแปลงคอลัมน์ที่เบ้มาก (|skew| > 1)
def transform_skewed_columns(df, numeric_columns):
    transformed_columns = []  # เก็บชื่อคอลัมน์ที่ถูกแปลง

    for col in numeric_columns:
        skewness = skew(df[col].dropna())  # คำนวณ skewness ของแต่ละคอลัมน์

        if abs(skewness) > 1:  # ถ้าเบ้เกิน 1 ถึงจะพิจารณาแปลง
            try:
                if (df[col] <= 0).any():
                    # ถ้ามีค่าติดลบ ใช้ Yeo-Johnson
                    df[col], _ = yeojohnson(df[col] + abs(df[col].min()) + 1)  # Shift ให้บวก
                    transformed_columns.append((col, 'Yeo-Johnson'))
                else:
                    # ถ้าทุกค่าบวก ใช้ Box-Cox
                    df[col], _ = boxcox(df[col] + 1)  # +1 เพื่อกัน log(0)
                    transformed_columns.append((col, 'Box-Cox'))
            except Exception as e:
                print(f"Error transforming column {col}: {e}")
                continue

    return transformed_columns

# ระบุคอลัมน์ที่เป็นตัวเลข
numeric_columns = df.select_dtypes(include=[np.number]).columns.tolist()

# เรียกใช้ฟังก์ชันแปลง
transformed_cols = transform_skewed_columns(df, numeric_columns)

# แสดงผลว่าคอลัมน์ไหนถูกแปลง และใช้วิธีอะไร
print("Transformed Columns:", transformed_cols)

# ถ้าต้องการย้อนกลับ ให้รันอันนี้
# df = df_backup.copy()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew

# สำรองข้อมูลคอลัมน์ก่อนแปลง
df['Num of pregnancies (backup)'] = df['Num of pregnancies'].copy().astype('float64')

# เช็ค Skewness ก่อนแปลง
skew_before = skew(df['Num of pregnancies'], nan_policy='omit')
print(f"Skewness ก่อนแปลง: {skew_before:.4f}")

# ลองแปลงด้วย log1p (log(1+x)) เพื่อจัดการเบ้
df['Num of pregnancies (log)'] = np.log1p(df['Num of pregnancies'])

# เช็ค Skewness หลังแปลง
skew_after = skew(df['Num of pregnancies (log)'], nan_policy='omit')
print(f"Skewness หลังแปลง (log1p): {skew_after:.4f}")

# ถ้าไม่ชอบ log จะลอง sqrt แทนก็ได้
df['Num of pregnancies (sqrt)'] = np.sqrt(df['Num of pregnancies'])

# เช็ค Skewness หลังแปลง (sqrt)
skew_after_sqrt = skew(df['Num of pregnancies (sqrt)'], nan_policy='omit')
print(f"Skewness หลังแปลง (sqrt): {skew_after_sqrt:.4f}")

# เปรียบเทียบ
print("\nเปรียบเทียบ Skewness")
print(f"Original: {skew_before:.4f}")
print(f"log1p: {skew_after:.4f}")
print(f"sqrt: {skew_after_sqrt:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# สร้างโมเดล
model = RandomForestClassifier(n_estimators=100, random_state=42)

# เทรนโมเดล
model.fit(X_train, y_train)

# ทำนายผลบนชุดทดสอบ
y_pred = model.predict(X_test)

# ประเมินผลลัพธ์
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# แสดง Confusion Matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# แสดง Precision, Recall, F1-score
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# ตรวจสอบค่า missing ในคอลัมน์ 'Age_group' และ 'Dx:Cancer'
missing_age_group = df['Age_group'].isna().sum()
missing_dx_cancer = df['Dx:Cancer'].isna().sum()

print(f"Missing values in 'Age_group': {missing_age_group}")
print(f"Missing values in 'Dx:Cancer': {missing_dx_cancer}")
# แปลงคอลัมน์ 'Age_group' ให้เป็น string (หากยังไม่เป็น)
df['Age_group'] = df['Age_group'].astype(str)

# ตรวจสอบค่าผิดพลาดใน 'Age_group' (เช่น ตรวจสอบว่ามีค่าที่ไม่เป็นกลุ่มอายุ)
invalid_age_groups = df[~df['Age_group'].str.contains(r'^\(\d{2},\s\d{2}\]$', na=False)]
print("Invalid Age Groups:")
print(invalid_age_groups)


# ตรวจสอบค่าผิดพลาดใน 'Dx:Cancer' (เช่น ตรวจสอบว่าไม่ควรมีค่าที่ไม่ใช่ 0 หรือ 1)
invalid_dx_cancer = df[~df['Dx:Cancer'].isin([0, 1])]
print("Invalid Dx:Cancer Values:")
print(invalid_dx_cancer)
```

```python
# คำนวณการเปรียบเทียบจำนวนการเป็นมะเร็งตามกลุ่มอายุ
age_group_cancer = df.groupby(['Age_group', 'Dx:Cancer']).size().unstack(fill_value=0)

# สร้างกราฟ heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(age_group_cancer, annot=True, fmt="d", cmap='Blues', cbar=False)

plt.title('Heatmap of Cancer by Age Group')
plt.xlabel('Dx:Cancer')
plt.ylabel('Age Group')
plt.show()
```
